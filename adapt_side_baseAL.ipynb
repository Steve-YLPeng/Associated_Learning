### side adapt 多次

```python
import gc
batch = 256
data = "imdb"
for test_count in range(10):
    text_len = 500
    for model in ["transformeralside","linearalside","lstmalside"]:
        layer = 5
        epoch = 50
        lr = 0.0005
        save_path = f"ckpt/{data}_{model}_l{layer}_{batch}_sidead/"
        out_path = f"result/{data}_{model}_l{layer}_{batch}_side{text_len}ad/"
        log = f"{out_path}/{data}_{model}_l{layer}_{batch}_sidead_{test_count}.log"
        !mkdir {save_path}
        !mkdir {out_path}
        !python3 dis_train_base_side.py --dataset {data} --model {model} --epoch {epoch} \
        --max-len {text_len} \
        --num-layer {layer}\
        --batch-size {batch} --lr {lr} --l1-dim 300 --label-emb 128 \
        --side-dim 100-100-100-100-100\
        --lr-schedule plateau \
        --save-dir {save_path} \
        --out-dir {out_path} \
        --task text > {log}
    gc.collect()
```
